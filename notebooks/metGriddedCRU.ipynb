```python
def getOneCell(met,x,y):
    noMonths = len(met)/360    #1200 # dividing by 360 gives the number of months in the grid file
    for i in range(10):
        mult = 360*i
        print(met[mult+x,y])        
```

```python
def getOneMonth(met,i,country=None):
    mult = 360*i
    month = met[mult:mult+360,:]
    month = np.flip(month,0)
    if country=='India':
        return month[100:170,490:560]
    else:
        return month
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

country = 'India'
# country = None

oneMonth = getOneMonth(met,120,country)
if country=='India':
    plt.figure(figsize=(10,10))
else:
    plt.figure(figsize=(36,18))

plt.imshow(oneMonth, interpolation='none')
plt.show()
```
