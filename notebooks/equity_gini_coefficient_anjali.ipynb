```python
import pandas
import glob
import os
```

```python
files = glob.glob('/Users/anjali/Downloads/Blocks/*.csv')

def calc_gini(file):
        experiment_name = file

        df=pandas.read_csv(experiment_name)
        df.columns
        dfa=df['a'].to_frame()
        dfa=dfa.dropna()
        dfa=dfa.rename(columns={'a':'col'})
        c = dfa

        # dfb=df['B'].to_frame()
        # dfb=dfb.dropna()
        # dfb=dfb.rename(columns={'B':'col'})
        # c=dfa.append(dfb)

        c.count()
        c=c.sort_values('col')
        c['ones']=1
        c
        c_cumsum=c['ones'].cumsum()
        c['cumsum_ones']=c_cumsum.to_frame()['ones']
        c['pct_cumsum_ones']=c['cumsum_ones']/c['cumsum_ones'].max()
        c_cumsum_col=c['col'].cumsum()
        c['cumsum_col']=c_cumsum_col.to_frame()['col']
        c
        c['pct_cumsum_col']=c['cumsum_col']/c['cumsum_col'].max()
        c
        c['diff_pct_cumsum']=c['pct_cumsum_ones']-c['pct_cumsum_col']
        c
        sum_c_diff_pct_cumsum=c['diff_pct_cumsum'].sum()
        sum_c_pct_cumsum_ones=c['pct_cumsum_ones'].sum()
        gini_coefficient=sum_c_diff_pct_cumsum/sum_c_pct_cumsum_ones

        return gini_coefficient, os.path.basename(experiment_name)

        
calc_gini('/Users/anjali/Downloads/Blocks/Aland1.csv')
        
for file in files:
    res = calc_gini(file)
    print(f"{res[0]},{res[1]}")
```
